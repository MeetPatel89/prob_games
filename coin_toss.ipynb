```python
import random
import matplotlib.pyplot as plt

# Define a function that tosses coin n times
def coin_toss(n):
    results = []
    for i in range(n):
        results.append(random.choice(["heads", "tails"]))
    return results
```

##### Toss coin 10 times

```python
results = coin_toss(10)
```

```python
plt.bar(["heads", "tails"], [results.count("heads"), results.count("tails")])
```

```python
results = coin_toss(100)
plt.bar(["heads", "tails"], [results.count("heads"), results.count("tails")])
```

```python
results = coin_toss(1000)
plt.bar(["heads", "tails"], [results.count("heads"), results.count("tails")])
```

```python
# Theoretical probability distribution of 10 fair coin flips 
# using pascal's triangle to get the distribution
outcomes = [10 for i in range(1)] + [9 for i in range(10)] + [8 for i in range(45)] + [7 for i in range(120)] + [6 for i in range(210)] + [5 for i in range(252)] + [4 for i in range(210)] + [3 for i in range(120)] + [2 for i in range(45)] + [1 for i in range(10)] + [0 for i in range(1)]
```

```python
arr = np.array(outcomes)
dict_dist = {i: outcomes.count(i) for i in arr}
print(dict_dist)

# use numpy.linspace to get symmetrical bin distribution
plt.hist(arr, density=True, bins=np.linspace(0, 11, num=12))
```
