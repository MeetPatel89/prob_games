```python
import pandas as pd
```

```python
die = pd.Series([1, 2, 3, 4, 5, 6])
# roll die 5 times
die.sample(n=5, replace=True)
```

```python
# write a function to get mean of n number of trials where each trial means rolling die 5 times
def get_sample_means(n):
    sample_means = []
    for i in range(n):
        sample_means.append(die.sample(n=5, replace=True).mean())
    return pd.Series(sample_means)
```

#### Simulate that as number of trials increases the distribution of sampling means tends towards normal distribution even though original random variable didn't have normal distribution. In this specific case initial random variable (outcome of die roll) has uniform discrete distribution. (Central Limit Theorem Experiment)

```python
get_sample_means(5).hist()
```

```python
get_sample_means(100).hist()
```

```python
get_sample_means(1000).hist()
```

```python
get_sample_means(10000).hist()
```
