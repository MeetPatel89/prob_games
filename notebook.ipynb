```python
from scipy.stats import uniform
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create an array from -3 to 8 with a difference of 0.1
x = np.arange(-3, 8, 0.1)
```

```python
# plot a uniform density function with 0 and 5 as lower and upper bounds respectively
y = uniform.pdf(x, 0, 5)
plt.plot(x, y)
plt.show()
```

#### Suppose a bus shows up at a bus stop every 20 minutes. If you arrive at the bus stop, what is the probability that the bus will show up in 8 minutes or less?

```python
uniform.cdf(8, 0, 20)
```

#### Experimental probability distribution for simulation of 1000 wait times for catching the bus

```python
wait_times = uniform.rvs(0, 20, size=1000)
plt.hist(wait_times, density=True)
```

#### The weight of a certain species of frog is uniformly distributed between 15 and 25 grams. If you randomly select a frog, what is the probability that the frog weighs between 17 and 19 grams?

```python
uniform.cdf(x=19, loc=15, scale=10) - uniform.cdf(x=17, loc=15, scale=10)
```
