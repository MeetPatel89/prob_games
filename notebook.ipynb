```python
from scipy.stats import uniform
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create an array from -3 to 8 with a difference of 0.1
x = np.arange(-3, 8, 0.1)
```

```python
# plot a uniform density function with 0 and 5 as lower and upper bounds respectively
y = uniform.pdf(x, 0, 5)
plt.plot(x, y)
plt.show()
```

#### Suppose a bus shows up at a bus stop every 20 minutes. If you arrive at the bus stop, what is the probability that the bus will show up in 8 minutes or less?

```python
uniform.cdf(8, 0, 20)
```
